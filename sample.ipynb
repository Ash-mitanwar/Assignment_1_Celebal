This function is created to clean numeric values by removing symbols such as ‚āĻ, commas, quotes and decimal points

```python
def fix_numb(x):
    if isinstance(x,str):
        return x.replace('‚āĻ', '').replace(',', '').replace('"', '').replace('.', '')
    return x
```

apply() is used for applying the function to a coulmn

```python
df['final_price'] =df['final_price'].apply(fix_numb)
```

```python
df
```

astype() is used to change the data type of the column

```python
df['final_price'] = df['final_price'].astype(int)
df['final_price'] = df['final_price']/100   
```
